## Análisis inicial (correlaciones e interpretativo)

Hay demasiadas columnas. Habiendo hecho una pequeña exploración previa por encima, haremos un análisis de covarianzas para reducir la cantidad de columnas que comienzas con 'tbp', para facilitar el EDA

```python
tbp_columns = df.filter(regex='^tbp').select_dtypes(include=['number'])

correlation_matrix = tbp_columns.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix of tbp Variables")
plt.show()
```

```python
threshold = 0.7

high_corr_pairs = correlation_matrix[(correlation_matrix.abs() > threshold) & (correlation_matrix != 1.0)]
high_corr_pairs = high_corr_pairs.stack().reset_index()
high_corr_pairs.columns = ['Feature1', 'Feature2', 'Correlation']
high_corr_pairs = high_corr_pairs.drop_duplicates()

high_corr_pairs.sort_values(by='Correlation', ascending=False)
```

```python
df_log = df.copy()
df_log['age_approx'] = df_log['age_approx'].transform(lambda x: np.log(x))
corrmat_log = df_log.corr()
```

```python
sns.heatmap(corrmat_log, center=0, cmap='jet', square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()

print(corrmat_log['target'])
```

```python
df_square = df.copy()
df_square = df_square.transform(lambda x: x**2)
corrmat_square = df_square.corr()
```

```python
sns.heatmap(corrmat_square, center=0, cmap='jet', square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()

print(corrmat_square['target'])
```

```python
df_squareroot = df.copy()
df_squareroot = df_squareroot.transform(lambda x: np.sqrt(x))
corrmat_squareroot = df_squareroot.corr()
```

```python
sns.heatmap(corrmat_squareroot, center=0, cmap='jet', square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()

print(corrmat_squareroot['target'])
```

```python
df_cubicroot = df.copy()
df_cubicroot = df_cubicroot.transform(lambda x: np.cbrt(x))
corrmat_cubicroot = df_cubicroot.corr()
```

```python
sns.heatmap(corrmat_cubicroot, center=0, cmap='jet', square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()

print(corrmat_cubicroot['target'])
```

```python
# Let's compare the results of all four transformations, and use the best ones
pd.DataFrame([corrmat['target'], corrmat_log['target'], corrmat_square['target'],
              corrmat_squareroot['target'], corrmat_cubicroot['target']],
             index=['', 'log', '^2', '^(1/2)', '^(1/3)'])
```

```python
df_transformed = df.copy()

df_transformed.tbp_lv_deltaLBnorm = df_square.tbp_lv_deltaLBnorm
df_transformed.tbp_lv_norm_border = df_square.tbp_lv_norm_border
df_transformed.tbp_lv_norm_color = df_square.tbp_lv_norm_color

df_transformed
```
